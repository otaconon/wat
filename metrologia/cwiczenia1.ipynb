## Zadanie 5

### Dane

```python
import pandas as pd

dane = {
  'Zawartość grafitu [% wag.]': [1, 2, 3, 4, 5, 6, 7],
  'Porowatość [%]': [0.2, 0.5, 0.9, 1.1, 1.6, 1.9, 2.2]
}

df = pd.DataFrame(dane)
```

### Tabela

```python
print("Pomiary wyprasek")
df
```

### Wykres punktowy

```python
plt.figure(figsize=(8, 6))

sns.regplot(
  x=df['Zawartość grafitu [% wag.]'], 
  y=df['Porowatość [%]'],
  scatter_kws={'color': 'blue'},
  line_kws={'color': 'red'},
  ci=None
)

plt.xlabel('Zawartosc grafitu [% wag.]')
plt.ylabel('Porowatosc [%]')
plt.title('Rys 1. Zależność zawartosci grafitu od porowatosci', y=-0.24)
plt.show()
```

### Metoda najmniejszych kwadratow

```python
x = df['Zawartość grafitu [% wag.]']
y = df['Porowatość [%]']

a, b = np.polyfit(x, y, 1)
print(f"Równanie prostej: y = {a:.4f}x + {b:.4f}")
```

```python
korelacja = df['Zawartość grafitu [% wag.]'].corr(df['Porowatość [%]'])
print(f"Współczynnik korelacji: {korelacja}")
```

```python
srednice = [
  49.00, 49.54, 50.00, 49.30, 50.10,
  49.51, 49.20, 50.02, 49.97, 50.08,
  50.18, 49.95, 49.91, 49.96, 49.91,
  49.87, 49.97, 50.05, 50.08, 49.85,
  49.80, 50.10, 49.90, 49.93, 50.00
]

n = len(srednice)
k = int(np.sqrt(n))

plt.figure(figsize=(10, 6))

n_bins, bins, patches = plt.hist(srednice, bins=k, color='skyblue', edgecolor='black', alpha=0.7)

plt.title(f'Histogram średnic wałków (n={n}, k={k})')
plt.xlabel('Średnica wałka [mm]')
plt.ylabel('Liczebność (częstość)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(np.round(bins, 2))
plt.show()

print("Granice przedziałów klasowych i liczebności:")
for i in range(len(n_bins)):
  print(f"Przedział {i+1}: {bins[i]:.2f} - {bins[i+1]:.2f} mm | Liczebność: {int(n_bins[i])}")
```
