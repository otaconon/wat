```python
import matplotlib.pyplot as plt
import numpy as np


plt.style.use("dark_background")
```

```python

# 1. Create a grid of x and y values
x_vals = np.linspace(-4, 4, 400)
y_vals = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x_vals, y_vals)

Z = X + 1j * Y

epsilon = 1e-9
W = np.abs((Z - 1j) / (Z + 1j + epsilon))

plt.figure(figsize=(6, 6))

contour = plt.contour(X, Y, W, levels=[2], colors='blue', linewidths=2)

plt.scatter([0], [1], color='red', label='i (Numerator)')
plt.scatter([0], [-1], color='green', label='-i (Denominator)')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.title('Direct Contour Plot of |z - i| / |z + i| = 2')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.show()
```
