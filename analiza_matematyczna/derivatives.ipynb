```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
plt.style.use("dark_background")
```

```python
def zad_58():
  def f(x):
    return 2/(x**(7/2))
  def df(x):
    return -7/(x**(9/2))
  
  x = np.linspace(0.6, 2, 100)
  y = f(x)
  dy = df(x)
  
  fig, ax = plt.subplots()  
  ax.plot(x, y)
  ax.plot(x, dy)
  ax.legend(["f(x)", "df/dx"])
  
  plt.show()


zad_58()
```
